```python
# Telegraaf .json to .csv

# Example URL to download json file: "https://www.telegraaf.nl/graphql?operationName=SearchPage&variables={%22first%22:4900,%22after%22:%22%22,%22hasSearchTerm%22:true,%22term%22:%22corona%22,%22ordering%22:null,%22sectionName%22:null}&extensions={%22persistedQuery%22:{%22version%22:1,%22sha256Hash%22:%226f1e3a18a82f910a76b155152fb36e8b5d0802658cf0586071b750daf9ecffd8%22}}"
# Has a 4900 result limit

import csv
import json

firstrow=["Medium", "Datum", "Kop", "URL"]
medium = "Telegraaf"
datum = ""
kop = ""
url = ""
count = 0

# Create csv
with open('Telegraafjson.csv', 'w', newline='', encoding="utf-8") as f:
    writer = csv.writer(f, delimiter = ";")
    writer.writerow(firstrow)

# Define writer
def csv_add(row):
    with open('Telegraafjson.csv', 'a', newline='', encoding="utf-8") as f:
        writer = csv.writer(f, delimiter = ";")
        writer.writerow(row)

t = open('Telegraaf4900 coronavirus.json', encoding='utf-8')
data = json.load(t)

count = 0 
subcount = 0

for a in data['data']['listArticlesBySearchString']:
#     print(a)
    url = a['url']
    datum = a['publishDate']
    kop = a['teaser']['title']
    nextrow = [medium, datum, kop, url]
    csv_add(nextrow)
    subcount+=1

    if subcount == 100:
        subcount=0
        count+=1
        print("Loaded: ", count*100)
    else:
        continue

print("Done.")
```

```python
# CSV VIA NEXIS
# Works for all search results in Nexis Uni.
# Requires Nexis Uni login credentials.
# Example URL: "	https://advance-lexis-com.proxy.uba.uva.nl/search/?pdmfid=1516831&crid=82a509fc-1c4b-4553-b6df-d1e90b90c70b&pdpsf=&pdpost=&pdstartin=&pdsearchterms=corona+or+coronavirus&pdsearchtype=SearchBox&pdtypeofsearch=searchboxclick&pdsf=MTA5NTk3OA~%5Enews~%5EDe+Telegraaf.nl&pdquerytemplateid=&pdtimeline=01%2F11%2F2020to01%2F12%2F2021%7Cdatebetween&pdfromadvancedsearchpage=true&ecomp=5xLg9kk&earg=pdpsf&prid=edec9e25-cfb9-4063-a322-907b04ae7ed3"

from datetime import datetime
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import os
import random
import time
import csv

medium = "NRC Nexis"
datum = ""
kop = ""
url = "Nexis"
firstrow=["Medium", "Datum", "Kop", "URL"]
print('.csv ready.')

# Create csv
with open('NRCNexis.csv', 'w', newline='', encoding="utf-8") as f:
    writer = csv.writer(f, delimiter = ";")
    writer.writerow(firstrow)

# Define writer
def csv_add(row):
    with open('NRCNexis.csv', 'a', newline='', encoding="utf-8") as f:
        writer = csv.writer(f, delimiter = ";")
        writer.writerow(row)

driver = webdriver.Chrome(executable_path="D:\Python\chromedriver.exe")

# URL HIER

driver.get("https://databases.uba.uva.nl/")
driver.maximize_window()
print("Webdriver started.")

nexis = driver.find_element_by_xpath("/html/body/div/div[2]/div/table/tbody/tr[350]/td[1]/a")
nexis.click()

time.sleep(20)
# LOGIN NOW

# SEARCH URL HERE
driver.get("https://advance-lexis-com.proxy.uba.uva.nl/search/?pdmfid=1516831&crid=4cc0199a-e437-4caa-89a0-2f75abd94c60&pdpsf=&pdpost=&pdstartin=&pdsearchterms=corona+or+coronavirus&pdsearchtype=SearchBox&pdtypeofsearch=searchboxclick&pdsf=MTA5NTg1MA~%5Enews~%5ENRC+Handelsblad.nl&pdquerytemplateid=&pdtimeline=01%2F02%2F2020to01%2F12%2F2021%7Cdatebetween&pdfromadvancedsearchpage=true&ecomp=5xLg9kk&earg=pdpsf&prid=fca82af7-fe84-4fd7-b3ca-abf9af4ada44")
print("Search results accessed.")

time.sleep(15)
# ADJUST SORT NOW

print("Starting scrape.")

# SCRAPING LOOP
count = 0
for x in range(1, 2000):
    
    time.sleep(5)
    
    source = driver.page_source
    soup = bs(source, 'lxml')
    
    count += 1
    print("Pages loaded: %s" % count)
    
    while True:
        results = soup.find_all("h2", class_="doc-title translate")
        if results == []:
            print("No results.")
            time.sleep(2)
            continue
        else:
            try:
                for item in results:
                        kop = item.text.replace(";"," ").replace("\n","")
                #             print("KOP ",kop)

                        ddlist = []
                        for dd in item.parent.find_all("dd"):
                            ddlist.append(dd)

                        datum = ddlist[4].text
                #             print("DATUM ",datum)

                        nextrow = [medium, datum, kop, url]
                        csv_add(nextrow)
                break
            except:
                print("ERROR: page ",count)
                page_source = driver.page_source
                with open('errorcache.txt', 'a', encoding='utf-8') as cache:
                    cache.write("\n")
                    cache.write(page_source)
                break
        
    while True:
        try:
            morebutton = driver.find_element_by_xpath("/html/body/main/div/main/div[2]/div/div[2]/div[2]/form/div[2]/nav/ol/li[11]")
            morebutton.click()
        except:
            try:
                morebutton = driver.find_element_by_xpath("/html/body/main/div/main/div[2]/div/div[2]/div[2]/form/div[2]/nav/ol/li[10]")
                morebutton.click()
            except:
                try:
                    morebutton = driver.find_element_by_xpath("/html/body/main/div/main/div[2]/div/div[2]/div[2]/form/div[2]/nav/ol/li[9]")
                    morebutton.click()
                except:
                    try:
                        morebutton = driver.find_element_by_xpath("/html/body/main/div/main/div[2]/div/div[2]/div[2]/form/div[2]/nav/ol/li[8]")
                        morebutton.click()
                    except:
                        try:
                            morebutton = driver.find_element_by_xpath("/html/body/main/div/main/div[2]/div/div[2]/div[2]/form/div[2]/nav/ol/li[7]")
                            morebutton.click()
                        except:
                            print("Button not clickable")
                            time.sleep(2)
                            continue
        break

print("Done.")
```

```python
# NRC .json to .csv
# Example URL: "https://www.nrc.nl/search/api/v1/article/search?article_type=alle_artikelen&date_range=specific_date&fromDate=2020-03-01&toDate=2021-12-01&query=corona%20coronavirus&from=0&size=12&aggregations%5B%5D=author&aggregations%5B%5D=categories"
# Limit in number of search results may require multiple .json files with different start and end dates.

import csv
import json

firstrow=["Medium", "Datum", "Kop", "URL"]
medium = "NRC"
datum = ""
kop = ""
url = ""
count = 0

# Create csv
with open('NRC10000 corona coronavirus vroeger 1.csv', 'w', newline='', encoding="utf-8") as f:
    writer = csv.writer(f, delimiter = ";")
    writer.writerow(firstrow)

# Define writer
def csv_add(row):
    with open('NRC10000 corona coronavirus vroeger 1.csv', 'a', newline='', encoding="utf-8") as f:
        writer = csv.writer(f, delimiter = ";")
        writer.writerow(row)

t = open('NRC10000 corona coronavirus vroeger 1.json', encoding='utf-8')
data = json.load(t)

count = 0 
subcount = 0

for a in data['hits']:
    kop = a['source']['headline']
    datum = a['source']['published_at']
    url = a['source']['url']
    nextrow = [medium, datum, kop, url]
    csv_add(nextrow)
    subcount+=1

    if subcount == 1000:
        subcount=0
        count+=1
        print("Loaded: ", count*1000)
    else:
        continue

print("Done.")
```
